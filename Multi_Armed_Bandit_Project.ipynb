```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bandits = [0.2, 0, -0.2, -5]

print("Bandits Created")
print(bandits)
```

```python
def pull_bandit(bandit):

    result = np.random.randn(1)

    if result > bandit:
        return 1
    else:
        return -1
```

```python
for i in range(10):

    choice = np.random.randint(4)

    reward = pull_bandit(bandits[choice])

    print("Bandit:", choice, "Reward:", reward)
```

```python
num_bandits = 4

total_reward = np.zeros(num_bandits)

print(total_reward)
```

```python
num_episodes = 1000

epsilon = 0.1

for episode in range(num_episodes):

    if np.random.rand(1) < epsilon:

        action = np.random.randint(num_bandits)

    else:

        action = np.argmax(total_reward)

    reward = pull_bandit(bandits[action])

    total_reward[action] += reward

print("Training Finished")
```

```python
for i in range(num_bandits):

    print(f"Bandit {i} Reward: {total_reward[i]}")
```

```python
best_bandit = np.argmax(total_reward)

print("Best Bandit is:", best_bandit)
```

```python
plt.bar(range(num_bandits), total_reward)

plt.xlabel("Bandits")
plt.ylabel("Rewards")
plt.title("Multi-Armed Bandit Rewards")

plt.show()
```
